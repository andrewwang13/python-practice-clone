<strong>Scrapes the Sierra Club Hundred Peaks Section website for info on the mountains on their list. (http://www.hundredpeaks.org/hpslist.htm)</strong>

```python
import csv
import requests
from BeautifulSoup import BeautifulSoup
import re
```

```python
base_url='http://www.hundredpeaks.org'
page_path='/hpslist.htm'
HPS_url=base_url+page_path
```

Parsing http://www.hundredpeaks.org/hpslist.htm

```python
response=requests.get(HPS_url) 
html=response.content
soup = BeautifulSoup(html)
```

Defining a function that goes into peak pages and pulls out elevation.

```python
def getElevation(url):
    response=requests.get(url)
    guide_html=response.content
    guide_soup=BeautifulSoup(guide_html)
    
    elev=[]
    
    try: 
        h2_list=guide_soup.findAll('h2')
        h2=h2_list[2]
        elev=h2.text
    except:
        elev='none'
          
    return elev
    
```

Defining a function that goes into peak pages and pulls out link to Caltopo map.

```python
def getCaltopo(url):
    response=requests.get(url)
    guide_html=response.content
    guide_soup=str(BeautifulSoup(guide_html))
    
    try:
        caltopo_id=re.search('(map\?id=\w*)',guide_soup)
        caltopo_url="http://caltopo.com/"+ caltopo_id.group(1)
    except:
        caltopo_url='na'
        
    return caltopo_url
```

Initiates and populates all_data with peak info.

```python
all_data=[]
all_data.append(['Peak','Guide URL','Elevation','Caltopo_url'])
```

```python
for i in soup.findAll('table',attrs={'cellpadding':6}):
#     print i
    for j in i.findAll('a'):
        row=[]
        row.append(j.text)
        guide_url=base_url+j.get('href')
        row.append(guide_url)
        elev=getElevation(guide_url)
        row.append(elev)
        row.append(getCaltopo(guide_url))
        all_data.append(row)
```
