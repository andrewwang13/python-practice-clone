```python
import unicodecsv as csv
import requests
from BeautifulSoup import BeautifulSoup
import random
import time
import re
```

```python
path_list=[]
with open('[EDU]_all_pages.csv','rU') as f:
    reader=csv.reader(f)
    path_list=list(reader)
```

```python
base_url = 'https://www.nerdwallet.com/'
url_list = []
for a in path_list:
    url_list.append(base_url + a[0])
    
url_list
```

```python
def find_phrase(url,phrase,output_list):
    response=requests.get(url)
    html=response.content
    soup=BeautifulSoup(html)
    
    body = soup.find('div',attrs={'property':'articleBody'})
    #print body
    per_story =[]
    if body != None:
        for j in body.findAll('p'):
            present = re.search(phrase,j.text)
            #print url
            row = []
            if present != None:
                counter=0
                row.append(url)
                row.append(present.group(0))
                row.append(j)
                for k in j.findAll('a'):

                    if re.search(present.group(0),k.text) != None:
                        counter+=1

                if counter > 0:
                    row.append('<a> tag DOES have phrase')
                else:
                    row.append('no <a> tag with phrase')

                output_list.append(row)
```
