This code scrapes the Census website to obtain a list that matches the NAICS number and name of sectors and subsectors in the 1997 NAICS.

```python
import csv
import requests
import pandas as pd
from BeautifulSoup import BeautifulSoup
```

Parsing main_url

```python
base_url='https://www.census.gov'
main_url='https://www.census.gov/eos/www/naics/reference_files_tools/1997/1997.html'
response=requests.get(main_url)
html=response.content
soup=BeautifulSoup(html)
```

Create list of subsector URLs

```python
all_data=[]
# Pull out sector URLs, make list
url_list=[]
table=soup.find('table',summary='The following table provides links to 1997 NAICS industry titles and descriptions.  Downloadable files are listed at the bottom of the page.')
```

```python
for i in table.findAll('a'):
	url_Path=i['href']
	url_list.append(base_url+url_Path)
```

```python
url_list
```

```python
for i in table.findAll('a'):
	row=[]
	row.append(i.text)
	trimmed_title=i['title'][16:]
	row.append(trimmed_title)
	all_data.append(row)
```

Define a function that goes into a section page and pulls out all subsector NAICS numbers and names

```python
def getSector(url):
	headings_list=[]
	sector_response=requests.get(url)
	sector_html=sector_response.content
	sector_soup=BeautifulSoup(sector_html)

	headings=sector_soup.findAll('h3')[4:]
	row=[]
	for i in headings:
		
		i_text=i.text
		row.append(i_text.split(" ",1))
	return row
```

Append those numbers/names to all_data

```python
for j in url_list:
	all_data=all_data+(getSector(j))
```

```python
all_data
```

Create CSV of all_data

```python
file = open("./1997_naics.csv","wb")
writer = csv.writer(file)
writer.writerows(all_data)
```
