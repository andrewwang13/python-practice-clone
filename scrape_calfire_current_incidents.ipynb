<strong>Pulling current fire information from Cal Fire's current incidents listing. (<a href="http://www.fire.ca.gov/current_incidents">http://www.fire.ca.gov/current_incidents)</a></strong>

```python
import requests
from BeautifulSoup import BeautifulSoup
import pandas as pd
```

```python
base_url='http://www.fire.ca.gov/current_incidents/incidentdetails/Index/'
```

Building list of valid incident URLs.

```python
incident_list=[]
counter=1700
while counter<=1900:
    incident_url=base_url+str(counter)
    response=requests.get(incident_url)
    
    if response.status_code==200:
        incident_list.append(incident_url)

    counter+=1
```

```python
all_data=[]
```

Building list of dictionaries to contain fire info.

```python
for i in incident_list:
    fire_response=requests.get(i)
    fire_html=fire_response.content
    fire_soup=BeautifulSoup(fire_html)
    
    fire_dict={}
    
    title=fire_soup.find('table',attrs={'id':'incident_information'})
    fire_dict['fire description']=title['summary']

    for i in fire_soup.findAll('tr',attrs={'class':'odd'}):
        key=i.findAll('td')[0]
        item=i.findAll('td')[1]
        fire_dict[key.text]=item.text
    
    all_data.append(fire_dict)
    
```

Data output.

```python
df=pd.DataFrame(data=all_data)
df
```
